```python
import random
class groups:
    def fun(self,team_list):
        group_1=[]
        group_2=[]
        for i1 in range(len(self.team_list)):
            x=random.choice(self.team_list)
            if (i1%2==0):
                group_1.append(x)
                self.team_list.remove(x)
            else:
                group_2.append(x)
                self.team_list.remove(x)
        df={}
        df["G1"]=group_1
        df["G2"]=group_2
        return df
    def matches_1(self,df):
        g1=[]
        g2=[]
        for i2 in range(5):
            for j2 in range(5):
                if i2!=j2:
                    g1.append(self.df['G1'][i2]+" vs "+self.df['G1'][j2])
                    g2.append(self.df['G2'][i2]+" vs "+self.df['G2'][j2])
        return g1+g2
    def matches_2(self,df):
        l3=[]
        for i3 in range(5):
            for j3 in range(5):
                if i3==j3:
                    l3.append(self.df['G1'][i3]+" vs "+self.df['G2'][j3])
                    l3.append(self.df['G2'][i3]+" vs "+self.df['G1'][i3])
        return l3
    def matches_3(self,df):
        l4=[]
        for i4 in range(5):
            c=i4
            for j4 in range(5):
                if i4!=j4:
                    if c%2==0:
                        l4.append(self.df['G1'][i4]+" vs "+self.df['G2'][j4])
                        c+=1
                    else:
                        l4.append(self.df['G2'][j4]+" vs "+self.df['G1'][i4])
                        c+=1
        return l4
teams=groups()
teams.team_list=["MI","CSK","RCB","DC","KKR","LSG","GT","RR","PK","SRH"]
df=teams.fun(teams.team_list)
print(df)
import pandas as pd
d=pd.DataFrame(df)
print(d)
teams.df=df
```

```python
t=teams.matches_1(df)
t1=teams.matches_2(df)
t2=teams.matches_3(df)
total=t+t1+t2
print(total)
print(len(total))
```
